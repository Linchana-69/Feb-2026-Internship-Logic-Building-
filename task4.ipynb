```python
def check_stock(products):
    for item, quantity in products.items():
        if quantity < 15:
            print(item, "Reorder Alert")
        else:
            print(item, "Stock OK ")
inventory = {
    "Rice": 20,
    "Sugar": 10,
    "Oil": 5,
    "Salt": 25
}

check_stock(inventory)
```

```python
def result(marks):
    total = 0
    for m in marks:
        total = total + m
    average = total / len(marks)
    if average >= 50:
        return "Pass"
    else:
        return "Fail"
marks = [40, 60, 50]
print(result(marks))
```

```python
def calculate_fare(distance, peak):
    fare = 50 + (12 * distance)

    if peak == "yes":
        fare = fare + (fare * 0.25)

    return fare


while True:
    try:
        distance = float(input("Enter distance in km: "))
    except ValueError:
        print("Please enter a valid number!")
        continue

    peak = input("Is it peak hour? (yes/no): ")

    total_fare = calculate_fare(distance, peak)
    print("Total Fare: ₹", total_fare)

    retry = input("Do you want to calculate again? (yes/no): ")

    if retry == "no":
        break
    elif retry != "yes":
        print("please enter valid ")
```

```python
def check_attendance(attendance):
    present = 0
    for day in attendance:
        if day == "P":
            present = present + 1
    percentage = (present / len(attendance)) * 100

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"
attendance_list = ["P", "P", "A", "P", "P"]
result = check_attendance(attendance_list)
print("Result:", result)
```

```python
def check_password(password):
    has_digit = False
    has_special = False
    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in "@#$":
            has_special = True
    if len(password) >= 8 and has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"
pwd = input("Enter password: ")
print(check_password(pwd))
```
