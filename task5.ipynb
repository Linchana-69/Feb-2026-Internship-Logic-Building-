```python
def analyze_engagement(likes_list):
    total_likes = 0


    for like in likes_list:
        total_likes += like

    print("Total Likes:", total_likes)


    if total_likes >= 1000:
        print("Post Status: Viral Post")
    else:
        print("Post Status: Normal Engagement")



likes = [200, 150, 300, 250, 150]   # Total = 1050
analyze_engagement(likes)
```

```python
def analyze_rainfall(rainfall_data, required_level):

    total_rainfall = 0


    for rain in rainfall_data:
        total_rainfall += rain


    average_rainfall = total_rainfall / len(rainfall_data)

    print("Average Rainfall:", int(average_rainfall))


    if average_rainfall >= required_level:
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")



rainfall = [70, 75, 68, 80, 67]
required = 70

analyze_rainfall(rainfall, required)
```

```python
def detect_duplicate_usernames(usernames):

    unique_usernames = set(usernames)


    if len(unique_usernames) != len(usernames):
        print("Duplicate Accounts Found: Yes")
        print("Status: Flag accounts")
    else:
        print("Duplicate Accounts Found: No")
        print("Status: Accounts are unique")



users = ["john123", "alice98", "mark77", "john123"]

detect_duplicate_usernames(users)
```

```python
def filter_premium_crops(prices):
    premium = []

    for price in prices:
        if price > 2000:
            premium.append(price)

    return premium



crop_prices = [1500, 2500, 1800, 3200, 2000]

premium_crops = filter_premium_crops(crop_prices)

print("Premium Crops:", premium_crops)
```

```python
def check_system_health(errors):
    if errors == 0:
        return "Healthy"
    elif errors <= 5:
        return "Minor Issues"
    else:
        return "Critical Issues"



error_count = 7
status = check_system_health(error_count)

print("Error Count:", error_count)
print("System Status:", status)
```

```python
def check_exam_eligibility(attendance_list):
    total_days = len(attendance_list)
    present_days = 0


    for day in attendance_list:
        if day == 1:      # 1 means Present
            present_days += 1

    attendance_percentage = (present_days / total_days) * 100

    if attendance_percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"

    return attendance_percentage, status



attendance = [1,1,1,1,0,1,1,1,1,0]  # 8 out of 10 present

percentage, eligibility = check_exam_eligibility(attendance)

print("Attendance Percentage:", percentage)
print("Exam Eligibility:", eligibility)
```

```python
def calculate_electricity_bill(units):
    bill = 0

    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)


    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"

    return bill, status



units_consumed = 250
total_bill, usage_status = calculate_electricity_bill(units_consumed)

print("Units Consumed:", units_consumed)
print("Total Bill: ₹", total_bill)
print("Usage Level:", usage_status)
```
