```python
class SmartLight:
    def __init__(self, name):
        self.name = name
        self.status = "OFF"   # Default status

    def turn_on(self):
        self.status = "ON"

    def turn_off(self):
        self.status = "OFF"

    def display_status(self):
        print(f"{self.name} is {self.status}")



light_name = input("Light Name: ")
action = input("Action (ON/OFF): ").upper()


light = SmartLight(light_name)

if action == "ON":
    light.turn_on()
elif action == "OFF":
    light.turn_off()
else:
    print("Invalid Action")


light.display_status()
```

```python
class Employee:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display(self):
        print(self.name)
        print(self.emp_id)
        print(self.department)


name = input("Employee Name: ")
emp_id = input("Employee ID: ")
department = input("Department: ")

emp = Employee(name, emp_id, department)


emp.display()
```

```python
class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def display(self):
        print("Contact Saved")
        print("Name:", self.name)
        print("Phone:", self.phone)



name = input("Contact Name: ")
phone = input("Phone Number: ")


contact1 = Contact(name, phone)


contact1.display()
```

```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def display_price_tag(self):
        print("Product:", self.name)
        print("Price: ₹" + str(self.price))



name = input("Product Name: ")
price = input("Price: ")


item = Product(name, price)


item.display_price_tag()
```

```python
class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    def display(self):
        print("Movie Name:", self.name)
        print("Rating:", self.rating + "/5")



name = input("Movie Name: ")
rating = input("Rating: ")


movie1 = Movie(name, rating)


movie1.display()
```

```python
class Delivery:
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address

    def display(self):
        print("Delivery Details")
        print("Customer:", self.customer_name)
        print("Address:", self.address)


name = input("Customer Name: ")
address = input("Address: ")


order = Delivery(name, address)

order.display()
```
