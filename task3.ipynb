```python
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}
higest_score = max(employees.values())
print("top performance eligible for bonus:")
for name in employees:
    if employees[name] == higest_score :
      print(name,"(score:",higest_score,")")
```

```python
query = "Buy mobile phone buy phone online"
query = query.lower()
words = query.split()
result = {}
for word in words:
    if words.count(word) > 1:
        result[word] = words.count(word)
print(result)
```

```python
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

valid_readings = []

for i in range(len(sensor_readings)):
    if sensor_readings[i] % 2 == 0:
        valid_readings.append((i, sensor_readings[i]))

print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

total = len(emails)

gmail = 0
yahoo = 0

for email in emails:
    if "gmail.com" in email:
        gmail += 1
    elif "yahoo.com" in email:
        yahoo += 1

print("gmail.com:", int((gmail / total) * 100), "%")
print("yahoo.com:", int((yahoo / total) * 100), "%")
```

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

avg = sum(sales) / len(sales)

for i in range(len(sales)):
    if sales[i] >= avg * 1.29:
        print("Day", i + 1, ":", sales[i])
```
